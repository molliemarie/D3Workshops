```python
import pandas as pd
```

```python
df = pd.read_csv('ChicagoCrimes-2001_to_present.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df['year'] = df['Date'].map(lambda x: x.year)
```

```python
df.head()
```

```python
grouped = df.groupby(['year','Primary Type']).size()
```

```python
grouped
```

```python
grouped.to_csv('ChiCrime.csv', header=['count'])
```
